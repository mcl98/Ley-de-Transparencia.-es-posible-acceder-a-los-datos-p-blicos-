```python
df = pd.read_csv('/content/drive/MyDrive/datosTFM/portal/portales.csv',sep=';', decimal=',',encoding='utf-8')
df
```

```python

# Normalizar la columna 'Cantidad' para que esté entre 0 y 1
df['Cantidad_normalizada'] = (df['Cantidad'] - df['Cantidad'].min()) / (df['Cantidad'].max() - df['Cantidad'].min())

# Escalar la columna normalizada para que esté entre 1 y 10
df['NotaCan'] = df['Cantidad_normalizada'] * 9 + 1

# Redondear la nota a un entero entre 1 y 10
df['NotaCan'] = df['NotaCan'].round(0).astype(int)

# Mostrar el dataframe con la nueva columna 'Nota'
df
```

```python

# Normalizar la columna 'Tematicas' para que esté entre 0 y 1
df['Tematicas_normalizada'] = (df['Tematicas'] - df['Tematicas'].min()) / (df['Tematicas'].max() - df['Tematicas'].min())

# Escalar la columna normalizada para que esté entre 1 y 10
df['NotaTem'] = df['Tematicas_normalizada'] * 9 + 1

# Redondear la nota a un entero entre 1 y 10
df['NotaTem'] = df['NotaTem'].round(0).astype(int)

# Mostrar el dataframe con la nueva columna 'Nota'
df
```

```python

# Normalizar la columna 'Formatos' para que esté entre 0 y 1
df['Formatos_normalizada'] = (df['Formatos'] - df['Formatos'].min()) / (df['Formatos'].max() - df['Formatos'].min())

# Escalar la columna normalizada para que esté entre 1 y 10
df['NotaFor'] = df['Formatos_normalizada'] * 9 + 1

# Redondear la nota a un entero entre 1 y 10
df['NotaFor'] = df['NotaFor'].round(0).astype(int)

# Mostrar el dataframe con la nueva columna 'Nota'
df
```

```python

# Normalizar la columna 'Porcentaje' para que esté entre 0 y 1
df['Porcentaje_normalizada'] = (df['Porcentajeac'] - df['Porcentajeac'].min()) / (df['Porcentajeac'].max() - df['Porcentajeac'].min())

# Escalar la columna normalizada para que esté entre 1 y 10
df['NotaPor'] = df['Porcentaje_normalizada'] * 9 + 1

# Redondear la nota a un entero entre 1 y 10
df['NotaPor'] = df['NotaPor'].round(0).astype(int)

# Mostrar el dataframe con la nueva columna 'NotaPor'
df
```

```python
# Cálculo de la media de todas las notas calculadas anteriormente

# Calcular la media de las notas calculadas anteriormente
df['NotaMedia'] = (df['NotaCan'] + df['NotaTem'] + df['NotaFor'] + df['NotaPor']) / 4

# Mostrar el dataframe con la nueva columna 'NotaMedia'
df
```

```python

# Ordenar el DataFrame por la columna 'NotaMedia' de mayor a menor
df_sorted = df.sort_values('NotaMedia', ascending=False)

# Mostrar el DataFrame ordenado
df_sorted
```
