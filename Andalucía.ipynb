```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/datosTFM/Andalucía/cobros.csv',sep='|', decimal=',',encoding='utf-8')
df
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Cálculo del porcentaje de datos erróneos asumiendo que estos son la suma de los datos faltantes y duplicados

# Cálculo del porcentaje de datos faltantes
missing_percentage = (df.isnull().sum() / len(df)) * 100
total_missing_percentage = missing_percentage.sum() / len(df.columns)


# Cálculo del porcentaje de datos duplicados
num_duplicates = df.duplicated().sum()
total_rows = len(df)
percentage_duplicates = (num_duplicates / total_rows) * 100


# Cálculo del porcentaje de datos incorrectos
total_incorrect_percentage = total_missing_percentage + percentage_duplicates

print("Porcentaje de datos incorrectos:", total_incorrect_percentage)
print("Porcentaje de datos faltantes:", total_missing_percentage)
print("Porcentaje de datos duplicaods:", percentage_duplicates)


def calculate_grade(percentage):
  """Calculates a grade from 1 to 10 based on the percentage.

  Args:
      percentage: The percentage value.

  Returns:
      An integer grade from 1 to 10.
  """
  grade = round(10 - (percentage / 100) * 9)
  return max(1, min(10, grade))  # Nota comprendida en el rango de  1-10


# Cálculo de la nota para cada porcentaje
grade_total_incorrect = calculate_grade(total_incorrect_percentage)
grade_total_missing = calculate_grade(total_missing_percentage)
grade_duplicates = calculate_grade(percentage_duplicates)


print("Nota precisión:", grade_total_incorrect)
print("Nota completitud:", grade_total_missing)
print("Nota unicidad:", grade_duplicates)
```
